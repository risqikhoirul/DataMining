```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_iris
import pandas as pd

# Load dataset iris sebagai contoh
iris = pd.read_excel("https://github.com/risqikhoirul/DataMining/raw/main/Iris%20-%20Semester%203.xlsx")
```

```python
iris.head()
```

```python
X = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = iris['Species']

# Inisialisasi model Decision Tree
model = DecisionTreeClassifier()
```

```python
# Fitting model dengan data
model.fit(X, y)
```

```python
# Lakukan K-Fold Cross Validation dengan K=5
kfold = 5
scores = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')
```

```python
# Tampilkan skor hasil cross validation
print(f"Accuracy: {scores.mean():.2f} (+/- {scores.std() * 2:.2f})")
```

```python
# Visualisasi Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(model, filled=True, feature_names=X.columns, class_names=iris['Species'].unique())
plt.show()
```

```python
# prediksi model dengan tree_model.predict([[SepalLength, SepalWidth, PetalLength, PetalWidth]])
print(model.predict([[6.2, 3.4, 5.4, 2.3]])[0])
```
